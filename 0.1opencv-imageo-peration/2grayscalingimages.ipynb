
<li>convert a color into grascale</li>
<li>see the changes between dimension between grayscale and color image</li>


```python
import cv2
from matplotlib import pyplot as plt
import numpy
```

```python
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread('./images/aquabeach.jpg')
image.shape
```

```python
imshow("aqua, beach", image, size = 5)
# imshow("aqua, beach", image)
```

```python
image.shape[:2] # (1200, 1920) (height, width)
def imshow(title = "", image = None, size = 10):
 
    # The line below is changed from w, h to h, w
    h, w = image.shape[0], image.shape[1]
    aspect_ratio = w/h
 
    plt.figure(figsize=(size * aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

imshow("aqua, beach", image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imshow("Converted to Grayscale", gray_image)
```

```python
cv2.imwrite('grey.jpg', gray_image)
```
